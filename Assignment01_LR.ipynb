```python
#Importing Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn import metrics
```

```python
#Read data from the source
df=pd.read_csv('/content/HousePrice.csv')
```

```python
df.head()
```

```python
df.isna().sum()
```

```python
df.info()
```

```python
df["Price ($)"] = df["Price ($)"].str.replace(",","").astype(int)
```

```python
df.head()
```

```python
#Segregating Independent and Dependent Variables
x=df['Size (sqft)'].values.reshape(-1,1)  #Independent Variable
y=df['Price ($)'].values.reshape(-1,1) #Dependent Variable
```

```python
print(x)
```

```python
#Building LR model
model=LinearRegression()
model.fit(x,y)
```

```python
#Make Predicctions
y_pred=model.predict(x)
print(y_pred)
```

```python
#Showing Bestfit line of regression
plt.scatter(df['Size (sqft)'],df['Price ($)'])
plt.plot(df['Size (sqft)'],y_pred,color='red')
plt.xlabel('Size (sqft)')
plt.ylabel('Price ($)')
plt.show()
```

```python
#coeficients
print(model.coef_)  #b1 Slope
print(model.intercept_) #b0 intercept
```

```python
#Model Evaluation
print("MSE:",metrics.mean_squared_error(y,y_pred))
print("RMSE:",np.sqrt(metrics.mean_squared_error(y,y_pred)))
print("R2:",metrics.r2_score(y,y_pred))
```

```python
#Predictions on User inputs
size=int(input("Enter size of plot: "))
print("Estimated Price(thousands):",model.predict([[size]]))
```
